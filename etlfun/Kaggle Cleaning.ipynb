```python
import pandas as pd
pd.set_option('display.max_columns', None)

from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
kaggle = pd.read_csv("./Kaggle.csv")
display(len(kaggle), kaggle.head())
```

```python
kaggle = kaggle[kaggle['adult'] == "False"]
kaggle = kaggle.drop('adult', axis=1)
kaggle['video'] = kaggle['video'].astype("boolean")
kaggle['budget'] = kaggle['budget'].astype('int64')
kaggle['id'] = kaggle['id'].astype('int64')
kaggle['popularity'] = kaggle['popularity'].astype('float64')
kaggle['release_date'] = kaggle['release_date'].astype('datetime64[ns]', errors = "raise")
display(len(kaggle), kaggle.dtypes, kaggle.head())
```
